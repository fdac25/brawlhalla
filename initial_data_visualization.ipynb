# Notebook For Displaying Early Versions Of Our Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("brawlhalla_matches.csv")

data
```

```python
# Diplaying Number Of Matches Per Tournament
# data.sort_values(by='date', inplace=True)
data['tournament_name'].value_counts().plot(kind='bar', figsize=(22, 6))
plt.xlabel('Tournament')
plt.ylabel('# Of Matches')
plt.show()
```

```python
# Diplaying Number Of Matches Per Event Type
data['event_name'].value_counts().plot(kind='bar', figsize=(10, 6))
plt.xlabel('Event')
plt.ylabel('# Of Matches')
plt.show()
```

```python
data['date'] = pd.to_datetime(data['date'])
data.sort_values(by='date', inplace=True)
data
```

```python
data['month'] = pd.to_datetime(data['date'], unit='s').dt.strftime('%Y-%m')
data
```

```python
monthly_counts = data.groupby('month').size().reset_index(name='matches')
monthly_counts
```

```python
plt.plot(monthly_counts['month'], monthly_counts['matches'])
plt.xticks(rotation='vertical')
plt.show()
```
